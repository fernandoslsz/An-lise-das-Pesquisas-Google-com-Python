```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
```

```python
tendencias = TrendReq(hl="pt-BR", tz=360)
tendencias.build_payload(kw_list=["pandemia"])
dados = tendencias.interest_by_region()
dados = dados.sort_values(by="pandemia", ascending=False)
dados = dados.head(10)
print(dados)
```

```python
dados.reset_index().plot(x="geoName", y="pandemia", 
                        figsize=(20,15), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```
